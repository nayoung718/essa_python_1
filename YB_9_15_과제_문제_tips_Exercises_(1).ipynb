### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

```python
data = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv", sep=',')
data
```

### Step 3. Assign it to a variable called tips

```python
tips=data
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop(index=0,axis=1)
tips
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips["total_bill"])
plt.xlabel("total_bill")
plt.ylabel("frequency")
plt.title("Histogram of total_bill")
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips.total_bill,tips.tip)
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.title("scatter plot (total_bill,tip)")
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
              x = tips['total_bill'],
              y = tips['tip'],
              s = tips['size']*10,
              c = tips['size']
)
axes1.set_title('relationship of total_bill, tip and size')
axes1.set_xlabel('total_bill')
axes1.set_ylabel("tip")
```

### Step 8. Present the relationship between days and total_bill value

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
              x = tips['day'],
              y = tips['total_bill'],
              c = tips['size']
)
axes1.set_title('day and total_bill')
axes1.set_xlabel('day')
axes1.set_ylabel("total_bill")
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x="tip", y="day", hue="sex", data=tips)
plt.xlabel("Tip")
plt.ylabel("Day")
plt.title("Tip and Day ")
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
plt.xlabel("day")
plt.ylabel("total_bill")
plt.title("total_bill per day (differentiated by time)")
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure()

plt.subplot(1, 2, 1)
sns.histplot(tips[tips["time"] == "Dinner"]["tip"])
plt.xlabel("tip")
plt.ylabel("frequency")
plt.title("dinner")

plt.subplot(1, 2, 2)
sns.histplot(tips[tips["time"] == "Lunch"]["tip"])
plt.xlabel("tip")
plt.ylabel("frequency")
plt.title("lunch")

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
tip_male = tips[tips["sex"] == "Male"]
tip_female = tips[tips["sex"] == "Female"]

plt.figure()

plt.subplot(1, 2, 1)
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tip_male)
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.title("Male")
plt.legend(title="Smoker")

plt.subplot(1, 2, 2)
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tip_female)
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.title("Female")
plt.legend(title="Smoker")

plt.tight_layout()
plt.show()
```
