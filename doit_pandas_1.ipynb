7. 박스그래프 그리기

```python
boxplot = plt.figure( )
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] =='Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
              labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

-다변량 그래프 그리기 : 3개 이상의 변수 사용한 그래프

-> 필요 이유: 지불금액 + 팁 + 성별 -> 점 색상 다르게 / 여기다 식사 비용 추가시 점 크기 다르게

-> 다변량 데이터로 다변량 그래프 그리기 : 산점도 그래프

1. 위의 산점도 그래프에 성별을 새 변수로 추가 : 색상으로 구분

-> female, male 같은 문자열은 0, 1 같은 정수 사용

```python
def recode_sex(sex):
  if sex =='Female':
    return 0
  else:
    return 1
```

2. recode_sex 메서드가 반환한 값(0,1)을 데이터프레임에 추가, sex_color 열 새로 생성하여 추가(sex 열에 recode_sex 함수를 브로드캐스팅하기 위해 apply 사용)

```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```

3. 테이블당 인원 수도 추가 : 점의 크기로 표현

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
  x=tips['total_bill'],
  y=tips['tip'],
  s=tips['size'] * 10,
  c=tips['sex_color'],
  alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill' )
axes1.set_ylabel('Tip')
```

##5. seaborn 라이브러리로 그래프 스타일 설정하기

1. 그래프에 스타일 적용하기

```python
fig, ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

2. whitegrid로 스타일 설정하여 그래프 그리기

```python
sns.set_style('whitegrid')
fig, ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

3. for문을 이용하여 모든 스타일을 하나씩 적용한 그래프

```python
fig = plt.figure( )
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)
fig.tight_layout()
```
